```python
import  pandas as pd 
df = pd.read_csv("tips.xls")
df.head()
```

```python
df.count()
```

```python
df.total_bill.max()
```

```python
%matplotlib inline
df.plot()
```

```python
df['total_bill'].plot(kind = 'hist')
```

```python
df['total_bill'].plot(kind = 'hist', grid = True, title = '  ')
```

```python
df['day'].value_counts().plot(kind = 'bar', 
                              grid = True, 
                              colormap = 'coolwarm', 
                              title = '   ')
```

```python
df.groupby(by = ['sex', 'smoker'])['tip'].mean().plot(kind = 'bar', grid= True)
```

```python
df[['total_bill', 'tip']].plot(kind = 'hist', 
                               grid = True, 
                               subplots = True, 
                               title = ['  ', ' '], 
                               legend = False)
```

```python
df.plot(x = 'total_bill', 
        y = 'tip', 
        kind = 'scatter', 
        grid = True, 
        title = '   Vs  ')
```

```python
df.pivot_table(values = ['total_bill', 'tip'], 
               index = 'day', 
               aggfunc = 'mean').plot(kind = 'bar')
```

```python
df.pivot_table(values = 'tip', index = ['sex', 'smoker'], aggfunc = 'count').plot(kind = 'bar')
```

```python
df.groupby('smoker')['tip'].mean().plot(kind = 'barh')
```

```python
df['smoker'].value_counts().plot(kind = 'bar')
```

```python
df.groupby('smoker')['tip'].count().plot(kind = 'bar')
```
